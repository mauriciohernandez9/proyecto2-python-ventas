```python
# Proyecto 2: Análisis de ventas de productos de salud
# Autor: Héctor Mauricio Hernández Olvera

import pandas as pd
import matplotlib.pyplot as plt

# Simulamos una base de datos
data = {
    'Producto': ['Termómetro', 'Guantes de Látex', 'Cubrebocas', 'Gel Antibacterial', 'Oxímetro'],
    'Unidades Vendidas': [1200, 850, 1500, 1000, 700],
    'Precio Unitario': [50, 1.5, 0.8, 2.0, 20]
}

# Convertimos a DataFrame
df = pd.DataFrame(data)

# Calculamos la venta total por producto
df['Venta Total'] = df['Unidades Vendidas'] * df['Precio Unitario']

# Mostramos tabla
print("Resumen de ventas por producto:")
print(df)

# Gráfico de barras
plt.figure(figsize=(10,6))
plt.bar(df['Producto'], df['Venta Total'], color='teal')
plt.title('Total de Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Venta Total ($)')
plt.xticks(rotation=30)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Producto más vendido
top_producto = df.loc[df['Venta Total'].idxmax()]
print(f"\nProducto más vendido: {top_producto['Producto']} con ${top_producto['Venta Total']:.2f}")
```
